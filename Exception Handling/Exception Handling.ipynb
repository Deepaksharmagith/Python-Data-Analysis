<span style="color:yellow; font-size:25px;">Exception key word</span>

```python
#Exception key word gives actual information about error.
try:
    a=input("Enter First Number")
    b= 10
    c=a/b
    print(c)
except Exception as e: #e is a variable, tell about error 
    print("Error Occured in Operation 1", e)
    print(type(e))  # telles type of erroe

try:
    print("Python Programming")
except Exception as h:
    print("Error Occured in Operation 2", h)
    print(type(h))
```

<span style="color:yellow; font-size:25px;">Use of finally: </span>

if program have a imp code of line which must be run, so we use finally block 
bez it execute the imp block either error ocurred in previous line of code.
This also decrese the line of code.

```python
f=open("demo-File.txt")
print("File is now open")
print(f.read())
f.close()
print("File is now closed")
```

```python
f=open("demo-File.txt")
print("File is now open") #--> program will terminate here further execution is not possible
print(f.red())
f.close()
print("File is now closed")
```

```python
try:
    f=open("demo-File.txt")
    print("File is now open")
    print(f.red())
except Exception as d:
    print(d)
    
finally:
    f.close()
    print("File is now closed")
```

```python
#Without using the final, line of code will increase.
try:
    f=open("demo-File.txt")
    print("File is now open")
except Exception as d:
    print(d)

try:
    print(f.red())
except Exception as h:
    print(h)

try:
    f.close()
    print("File is now closed")
except Exception as t:
    print(t)   
```
